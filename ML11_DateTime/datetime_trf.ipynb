```python
import pandas as pd
import numpy as np
```

<h3> Working with dates

```python
orders = pd.read_csv('orders.csv')
orders.head()
```

```python
orders.info()
```

```python
orders['date'] = pd.to_datetime(orders['date'] )


```

```python
orders.info()
```

```python
orders['date_year']= orders['date'].dt.year
```

```python
orders.head()
```

```python
orders['month']= orders['date'].dt.month
```

```python
orders.head()
```

```python
orders['day']= orders['date'].dt.day
```

```python
orders.head()
```

```python
orders.info()
```

```python
orders['day_name']= orders['date'].dt.day_name()
orders.head(4)
```

```python
orders['month_name'] = orders['date'].dt.month_name()
orders.head(4)
```

```python
# whether it is weekend or not

orders['is_weekend'] = np.where(orders['day_name'].isin(['Saturday','Sunday']), 1,0)
orders.head(4)
```

```python
# extracting the number of the quarter based on the date

orders['quarter'] = orders['date'].dt.quarter
orders.head()
```

```python
# to divide the quarter into semesters
orders['semester'] = np.where(orders['quarter'].isin([1,2]),1,2)

orders.head()
```

```python
# extract collapsed date
import datetime

today = datetime.datetime.today()
today
```

```python
(today-orders['date']).dt.days.head(4)
```

```python
np.round((today -orders['date']) / np.timedelta64(1, 'W'),0)
```
