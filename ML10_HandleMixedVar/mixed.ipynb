```python
fig = df['number'].value_counts().plot(kind='bar')
fig.set_title('No. of Passengers travelling together')
```

```python
df['number_numeric'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
```

```python
# np.where(condition, x, y), the order is always:
'''Condition: The "Question" (Is this null?)
x: The "Yes" answer (What to do if it is null)
y: The "No" answer (What to do if it is NOT null)'''

df['number_category' ] = np.where(df['number_numeric'].isnull(), df['number'], np.nan)
```

```python
df.head(5)
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # captures numerical part
df['cabin_cat'] = df['Cabin'].str[0] # captures the first letter

df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(10)
```
