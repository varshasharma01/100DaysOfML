```python
import pickle
import numpy as np
```

```python
ohe_gender = pickle.load(open('models/sex_ohe.pkl', 'rb'))
ohe_embarked = pickle.load(open('models/embarked_ohe.pkl','rb'))
classifier = pickle.load(open('models/classifier.pkl','rb'))
```

```python
test_input = np.array([1, 'female', 30.0, 1, 0, 17.6, 'C'], dtype=object).reshape(1,7)
```

```python
test_input
```

```python
test_input_gender = ohe_gender.transform(test_input[:,1].reshape(1,1))
```

```python
test_input_gender
```

```python
test_input_gender
```

```python
test_input_embarked = ohe_embarked.transform(test_input[:,-1].reshape(1,1))
```

```python
test_input_age = test_input[:,2].reshape(1,1)
```

```python
test_input_transformed = np.concatenate((test_input[:, [0,3,4,5]],test_input_age,test_input_gender , test_input_embarked), axis =1)
```

```python
classifier.predict(test_input_transformed)
```
